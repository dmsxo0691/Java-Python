```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

page = requests.get("https://news.nate.com/recent?mid=n0100")

soup = bs(page.text, "html.parser")

elements = soup.select('div.mlt01 a.lt1')

titles = []
links = []

for index, element in enumerate(elements, 1):
    titles.append(element.text)
    links.append("https:" + element.attrs['href'])

df = pd.DataFrame()
df['titles'] = titles
df['links'] = links

df.to_excel('./news.xlsx', sheet_name='Sheet1')
```
