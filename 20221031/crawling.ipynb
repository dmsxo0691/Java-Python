```python
import requests
from bs4 import BeautifulSoup
url = "https://www.daum.net"
req = requests.get(url)
data = req.text
print(data)
```

```python
# requests 라이브러리 이용해서 html 정보 가져오기
htmls = requests.get('https://www.daum.net/')

# BeautifulSoup 이용하여 파싱
bs = BeautifulSoup(htmls.content, 'html.parser')

print(bs)
```

```python

title = bs.find('title')

print(title)
```

```python
title = bs.head.title
print(title)
```

```python
url = "https://search.daum.net/search"
param = {'q':'python'}
req = requests.get(url.param = param)
data = req.text
print(data)
```
