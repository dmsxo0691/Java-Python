```python
is_seoul = data['지점명'] == '서울'
seoul = data[is_seoul]
seoul
```

```python
busan = data[data['지점명'] == '부산']
busan
```

```python
# 전처리 (날짜, 개수 24개 체크)
temp = []
for i in seoul['일시']:
    temp.append(i.split(' ')[0])
temp
```

```python
temp = []
for i in seoul['일시']:
    temp.append(i.split(' ')[0])
lack_seoul = []
unique_temp = set(temp)
unique_temp
```

```python
temp = []
for i in seoul['일시']:
    temp.append(i.split(' ')[0])
lack_seoul = []
unique_temp = set(temp)
for i in unique_temp:
    if temp.count(i) != 24:
        lack_seoul.append(i)
print(lack_seoul)
```

```python
temp = []
for i in busan['일시']:
    temp.append(i.split(' ')[0])
lack_busan = []
unique_temp = set(temp)
for i in unique_temp:
    if temp.count(i) != 24:
        lack_busan.append(i)
print(lack_busan)
```

```python
lack_data = list(set(lack_seoul+lack_busan))
lack_data
```

```python
for i in lack_data:
    seoul = seoul[seoul['일시'].str.split(' ').str[0] != i] # 서로 누락된 데이터가 있는 정보/중복되는 시간 제외
    busan = busan[busan['일시'].str.split(' ').str[0] != i]
```
