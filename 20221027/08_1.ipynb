```python
import matplotlib.pyplot as plt
result = pd.read_csv('C:\\Users\\KGA_09\\Downloads\\d1.csv',encoding='cp949')
data = {}
for i in range(1,len(result)):
    data[result.iloc[i,0]]= result.iloc[1, range(1,len(result.columns),6)]

plt.ylim([40000,50000])
for i in range(len(data['세종특별자치시  조치원읍(3611025000)'])):
    data['세종특별자치시  조치원읍(3611025000)'][i] = int(data['세종특별자치시  조치원읍(3611025000)'][i].replace(',',''))
    plt.plot(data['세종특별자치시  조치원읍(3611025000)'].index[i].split('년_')[0],data['세종특별자치시  조치원읍(3611025000)'][i],'ro')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:\\Users\\KGA_09\\AppData\\Local\\Microsoft\\Windows\\Fonts\\netmarble\\netmarbleM.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
import pandas as pd
result = pd.read_csv('C:\\Users\\KGA_09\\Downloads\\d1.csv',encoding='cp949')
data = {}
for i in range(1,len(result)):
    data[result.iloc[i,0]]= result.iloc[1, range(i,len(result.columns),6)]

name = input('세종특별자치시의 행정구역을 입력해주세요: ') 
for i,j in data.items():
    if i.find(name) != -1:
        for k in range(len(data[i])):
            data[i][k] = int(data[i][k].replace(',',''))
            plt.bar(data[i].index[k].split('년_')[0],data[i][k])
        plt.ylim([min(data[i])-3000,max(data[i])+3000])
plt.show()
```
